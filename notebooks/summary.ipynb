# Real Estate Industry Project Summary

## The overall approach of our project is as the following: 

1.We first srape some properties data on the https://www.domain.com.au/, and then preprocess the data.The data includes the weekly rental price of each property,which type of property it is, the number of bedrooms/bathrooms/parking spot it has, etc.

2.Then based on the location of these properties, we use Open Route Service API to get some geospatial data, which includes the distance of each property to Melbourne CBD, the number of schools, parks, supermarkets and cinemas within the scope of 5 minutes car-driving near each property.

3.Obtainig the cencus data which is from https://www.abs.gov.au/, we get the yearly personal income data and population data by SA2 districts. After forecasting the future population and applying a conversion of SA2 districts to postcode area, we combine the data with all the required information together.

4.Then to determine which features are essential on rental prices, we use heatmap to visualize the Pearson correlation coefficients of our obtained features,and it is showing below.

From this correlation heatmap, we obtained that the rental bond, number of bedrooms, number of bathrooms and number of parking spot are the most important internal features; while the number of nearby schools, the number of nearby cinemas, and its distance to Melbourne CBD are three most important external features.

5.Based on the results of the Pearson correlation coefficients of each features, we drop some features which are considered as having low effect on predicting rental prices. And finally, the features which we are using to model the data are: type, rental bond, distance to CBD, the number of bedrooms, the number of bathrooms, the number of parking spot,the number of nearby schools, the number of nearby parks, the number of nearby cinemas, if the property has air conditioning, if the property hasensuite, if the property has heating, if the property is furnished, if the property is close to shops, and if the property is close to transport.

6.We have tried fitting various models to forecast the rental prices,then based on our evaluation metrics which are R-squared value and Mean Squared Error to determine our final model. First, three machine learning models include Linear Regression,Random Forest and Decision Tree are implemented, then due to the limited API quota, our data is somewhat lacking, ensembling learing technique includes bagging and boosting is applied. Gradient Boosting Decision Tree, XGBoost, Bagging Decision Trees are used to fit the data, and comparing the overall result, we determine that, the bagging based model has the best performance among all the models with a R-squared value of 0.87 and MSE of 0.03.
